```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('Summary VB Stats UOregon 2019-2023.xlsx')
data.columns = data.iloc[0]
data = data.drop(0)

#data.columns
data.columns = ['#', 'Last Name', 'Position', 'Points', 'GP', 'GP%', 'Attack_Att', 'Ks', 'Ks%', 'Eff', 'Es%', 
                'Bs%', 'Pass_Att', 'Ind3', 'Ind4', 'GP%', 'FBSO%', 'Serve_Att', 'Aces%', 'Err', 'Err%', 
                'PS%', 'STFs', 'T%', 'GT%', 'E%', 'Touched', 'Digs', 'Digs%', 'CRT%', 'CNT%']

data
#data.columns
```

## Points comparison from 2019-2023:

```python
# column_values = data['Last Name '].unique()
# print(column_values)

ore_pts = data.loc[data['Last Name']=='ORE ', 'Points'].values[0]
plt.bar(['Oregon'], [ore_pts], label='Oregon', color='mediumpurple')

opp_pts = data.loc[data['Last Name']=='OPPONENT(S) ', 'Points'].values[0]
plt.bar(['Opponents'], [opp_pts], label='Opponents', color='pink')

plt.text(0, ore_pts, str(ore_pts), ha='center', va='bottom')
plt.text(1, opp_pts, str(opp_pts), ha='center', va='bottom')

plt.title('Total points comparison from 2019-2023')
plt.ylabel('Points')
plt.xlabel('Team')
plt.show()
```

# Offensive stats:

```python
off_stats = data[['#', 'Last Name', 'Position', 'Points', 'GP','Attack_Att', 'Ks', 'Ks%', 'Eff', 'Es%', 'Bs%', 
                  'Serve_Att', 'Aces%', 'Err', 'Err%', 'PS%']]

off_stats
```

```python
eff = off_stats['Eff'].values[off_stats['Eff'] != 0]
print(f"The average attack efficiency of the team is: {np.mean(eff[:-2])*100:.2f}%.")
aces = off_stats['Aces%'].values[off_stats['Aces%'] != 0]
print(f"The average ace percentage of the team is: {np.mean(aces[:-2]):.2f}%.")
att_err = off_stats['Es%'].values[off_stats['Es%'] != 0]
print(f"The average attack error of the team is: {np.mean(att_err[:-2]):.2f}%.")
ser_err = off_stats['Err%'].values[off_stats['Err%'] != 0]
print(f"The average serving error of the team is: {np.mean(ser_err[:-2]):.2f}%.")
ps = off_stats['PS%'].values[off_stats['PS%'] != 0]
print(f"The average point-scoring of the team is: {np.mean(ps[:-2]):.2f}%.")
```

## Kill graph:

```python
last_name = off_stats['Last Name'].values[off_stats['Ks'].values != 0]
kills = off_stats['Ks'].values[off_stats['Ks'] != 0][:-2]
plt.bar(last_name[:-2], kills)
plt.xticks(rotation=30, ha='right')
plt.axhline(y=np.mean(kills), color='red', linestyle='--', label=f'Avg. Kill Count: {np.mean(kills):.2f}')
plt.legend()
plt.show()
```

```python
#serve_att vs aces:
last_name_ser_att = off_stats['Last Name'].values[off_stats['Serve_Att'] != 0][:-2]
last_name_ace = off_stats['Last Name'].values[off_stats['Serve_Att'] != 0][:-2]
ser_att = off_stats['Serve_Att'].values[off_stats['Serve_Att'] != 0][:-2]
ser_att_percent = (ser_att / ser_att.sum()) * 100
# Filter out zero Serve Attempts for Aces
aces = off_stats['Aces%'].values[off_stats['Serve_Att'] != 0][:-2]
bar_width = 0.35
x_ser_att = np.arange(len(last_name_ser_att))
x_ace = np.arange(len(last_name_ace))
plt.bar(x_ser_att - bar_width/2, ser_att_percent, width=bar_width, label='Serve Attempts', color='pink')
plt.bar(x_ace + bar_width/2, aces, width=bar_width, label='Aces', color='olive')
plt.xticks(np.arange(len(last_name_ser_att)), last_name_ser_att, rotation=30, ha='right')
plt.xlabel('Players')
plt.ylabel('% Frequency')
plt.legend()
plt.show()
```

## Ace graph:

```python
last_name = off_stats['Last Name'].values[off_stats['Aces%'].values != 0]
aces = off_stats['Aces%'].values[off_stats['Aces%'] != 0][:-2]
plt.bar(last_name[:-2], aces)
plt.xticks(rotation=30, ha='right')
plt.axhline(y=np.mean(aces), color='red', linestyle='--', label=f'Avg. Ace %: {np.mean(aces):.2f}')
plt.legend()
plt.show()
```

# Defensive stats:

```python
def_stats = data[['#', 'Last Name', 'Position', 'Points', 'GP', 'STFs', 'T%', 'GT%', 'E%', 'Touched', 'Digs', 
                  'Pass_Att','Digs%', 'CRT%', 'CNT%']]

def_stats
```

```python
b_stuffs = def_stats['STFs'].values[def_stats['STFs'].values != 0]
print(f"The average block stuff count of the team is: {np.mean(b_stuffs[:-2]):.2f}.")
b_touch = def_stats['T%'].values[def_stats['T%'].values != 0]
print(f"The average block touch of the team is: {np.mean(b_touch[:-2]):.2f}%.")
good_b_touch = def_stats['GT%'].values[def_stats['GT%'].values != 0]
print(f"The average good block touch of the team is: {np.mean(good_b_touch[:-2]):.2f}%.")
digs = def_stats['Digs'].values[def_stats['Digs'].values != 0]
print(f"The average dig count per player is: {np.mean(digs[:-2]):.2f}.")
#important:
crt = def_stats['CRT%'].values[def_stats['CRT%'].values != 0]
print(f"The average dig count that resulted in an attack per player is: {np.mean(crt[:-2]):.2f}%.")
cnt = def_stats['CNT%'].values[def_stats['CNT%'].values != 0]
print(f"The average dig count that resulted in a point per player is: {np.mean(cnt[:-2]):.2f}%.")
#print(f"{np.mean(def_stats['Digs%']):.2f}") #=74.95
```

## Block graph:

```python
last_name = def_stats['Last Name'].values[def_stats['STFs'].values != 0]
plt.bar(last_name[:-2], b_stuffs[:-2])
plt.xticks(rotation=30, ha='right')
plt.axhline(y=np.mean(b_stuffs[:-2]), color='red', linestyle='--', label=f'Avg. Block Stuff Count: {np.mean(b_stuffs[:-2]):.2f}')
plt.legend()
plt.show()
```

## Dig count resulting in point graph:

```python
last_name = def_stats['Last Name'].values[def_stats['CNT%'].values != 0]
plt.bar(last_name[:-2], cnt[:-2])
plt.xticks(rotation=30, ha='right')
plt.axhline(y=np.mean(cnt[:-2]), color='red', linestyle='--', label=f'Avg. Dig Count % that Resulted in Point:{np.mean(cnt[:-2]):.2f}%')
plt.legend()
plt.show()
```

```python
#dig vs dig resulting in a point:
last_name_dig = def_stats['Last Name'].values[def_stats['Digs%'] != 0][:-2]
last_name_cnt = def_stats['Last Name'].values[def_stats['Digs%'] != 0][:-2]
dig = def_stats['Digs%'].values[def_stats['Digs%'] != 0][:-2]
# Filter out zero Serve Attempts for CNT
cnt = def_stats['CNT%'].values[def_stats['Digs%'] != 0][:-2]
bar_width = 0.35
x_dig = np.arange(len(last_name_dig))
x_cnt = np.arange(len(last_name_cnt))
plt.bar(x_dig - bar_width/2, dig, width=bar_width, label='Dig %', color='paleturquoise')
plt.bar(x_cnt + bar_width/2, cnt, width=bar_width, label='Dig resulting in point %', color='mediumslateblue')
plt.xticks(np.arange(len(last_name_dig)), last_name_dig, rotation=30, ha='right')
plt.xlabel('Players')
plt.ylabel('% Frequency')
plt.legend()
plt.show()
```
